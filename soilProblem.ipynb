```python
# dataset file path
file_path = '/content/TARP.csv'
# Laod from the dataset
names = ['Soil Moisture', 'Temperature', ' Soil Humidity', 'Time',
         'Air temperature (C)', 'Wind speed (Km/h)', 'Air humidity (%)',
         'Wind gust (Km/h)', 'Pressure (KPa)', 'ph', 'rainfall', 'N', 'P', 'K',
         'status'
        ]
irrigation_df = pd.read_csv(file_path, names=names, header=0)
```

```python
# Take a peek at the dataset
print(irrigation_df.shape)
irrigation_df.sample(5)
# N of rows
print("Rows: ",irrigation_df.shape[0])

# Column names
print("Columns: ",irrigation_df.columns.tolist())

# Data types
print("Data Type: \n",irrigation_df.dtypes)
```

```python
data = irrigation_df
stats_df = data.describe() # The describe() method generates descriptive statistics for the numeric columns in the DataFrame
stats_df.loc['range'] = stats_df.loc['max'] - stats_df.loc['min']
out_fields = ['mean', '25%', '50%', '75%', 'range', 'std']
stats_df = stats_df.loc[out_fields]
stats_df.rename({'50%': 'median'}, inplace=True)
stats_df

# Median and Mean Comparison:
# When the median and mean are close to each other
# (like in the case of "Soil Moisture"), the distribution is likely symmetric.
# However, when the mean is far from the median (like for "Temperature"), the distribution may be skewed.
# Range and Std:
# A large range combined with a moderate standard deviation can indicate
# that the data has outliers or is spread across a wide range,
# but the majority of the values are still clustered around the mean.
```

```python
# due to insufficient equippement we had to dropp all these columns
columns_to_drop = ['Wind gust (Km/h)', 'Pressure (KPa)', 'ph', 'rainfall' , 'N' , 'P' , 'K' , 'Time']
data.drop(columns=columns_to_drop, axis=1, inplace=True)
data.head(6)
```

```python
# searching for none values
check_for_nan = data.isnull().values.any()
sum_nan = data.isnull().sum().sum()
sum_nan_col = data.isnull().sum()
print("NaN Exist ? ", check_for_nan)
print("NaN Sum : ", sum_nan)
print("NaN Sum By Columns: \n", sum_nan_col)
```

```python
# changing status into binary
def status_to_binary(status):
    if status == 'ON':
        val = 1
    elif status == 'OFF':
        val = 0
    return val
data['status'] = data['status'].apply(status_to_binary)
data.head(6)
```

```python
# Dropping all NaN values
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame)
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(axis=1)
    return df[indices_to_keep].astype(np.float64)

new_data = clean_dataset(data)
new_data.shape # we will display the final shape that has no nan values
```

```python
# status On / Off
ax = sns.countplot(x="status", hue="status", data=data)
# calculate percetage
def calc_percentage(plot, feature):
    total = len(feature)
    for p in plot.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), size = 12)
    plt.show()
calc_percentage(ax, data.status)
```

```python
# 6 feautures
X = new_data.iloc[:, :-1].values
print(X.shape)
# status
y = new_data["status"].values
print(y.shape)
```

```python
# splitting dataset Train, Test (80%, 20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11111)
```

```python
# First let's normalize the data
# This aids the training of neural networks by providing numerical stability

normalizer = StandardScaler()
X_train_norm = normalizer.fit_transform(X_train)
X_test_norm = normalizer.transform(X_test)
```

```python
model = Sequential()
model.add(Flatten(input_shape=(6,)))
model.add(Dense(24,  activation="sigmoid"))
model.add(Dense(24,  activation="sigmoid"))
model.add(Dense(1, activation="sigmoid"))
model.compile(optimizer=SGD(learning_rate=1e-3), loss="binary_crossentropy", metrics=["accuracy"])
run_hist = model.fit(X_train_norm, y_train, validation_data=(X_test_norm, y_test), epochs=500)
```

```python
# Evaluate the model on the test data to check the accuracy
test_loss, test_accuracy = model.evaluate(X_test_norm, y_test)

# Print the test accuracy
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=2)  # You can choose the number of neighbors (k)

# Train the KNN model
knn.fit(X_train_norm, y_train)
```

```python
# Evaluate the KNN model on the test data
test_accuracy = knn.score(X_test_norm, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Additional performance metrics
y_pred = knn.predict(X_test_norm)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
# Initialize the Decision Tree classifier
dt = DecisionTreeClassifier(random_state=11111)

# Train the Decision Tree model
dt.fit(X_train_norm, y_train)
```

```python
# Evaluate the Decision Tree model on the test data
test_accuracy = dt.score(X_test_norm, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Additional performance metrics
y_pred = dt.predict(X_test_norm)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
# Initialize the Random Forest classifier
rf = RandomForestClassifier(n_estimators=100, random_state=11111)  # You can adjust n_estimators (number of trees)

# Train the Random Forest model
rf.fit(X_train_norm, y_train)
```

```python
# Evaluate the Random Forest model on the test data
test_accuracy = rf.score(X_test_norm, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Additional performance metrics
y_pred = rf.predict(X_test_norm)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
# Initialize the Linear Regression model
lr = LinearRegression()

# Train the Linear Regression model
lr.fit(X_train_norm, y_train)
```

```python
# Predict the target values using the trained model
y_pred = lr.predict(X_test_norm)

# Convert the continuous predictions to binary (e.g., 0 or 1)
y_pred_binary = [1 if i >= 0.5 else 0 for i in y_pred]

# Evaluate the model using accuracy
test_accuracy = accuracy_score(y_test, y_pred_binary)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Additional performance metrics
print(classification_report(y_test, y_pred_binary))
```
